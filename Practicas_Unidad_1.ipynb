```python
import numpy as np

# Creamos un array de una sola dimensión (vector)
vector = np.array([1, 2, 3, 4, 5])

# Creamos un array de dos dimensiones (matriz de 2 filas y 3 columnas)
matriz = np.array([[1, 2, 3], [4, 5, 6]])

# Mostramos las propiedades del vector:
# .ndim: Número de dimensiones (para un vector es 1)
# .shape: Dimensiones exactas (en este caso, 5 elementos en el primer eje)
print(vector.ndim, vector.shape)

# Mostramos las propiedades de la matriz:
# .ndim: Número de dimensiones (para una matriz es 2)
# .shape: Tupla con (filas, columnas), aquí sería (2, 3)
print(matriz.ndim, matriz.shape)
```

***Operaciones aritméticas y ufuncs***

```python
import numpy as np

# Definimos dos matrices cuadradas de 2x2
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Suma de matrices: se suma cada elemento con su correspondiente (1+5, 2+6, etc.)
C = A + B

# Operación escalar: calcula la raíz cuadrada de cada elemento dentro de la matriz A
D = np.sqrt(A)

# Imprime el resultado de la suma (Matriz 2x2)
print("Suma de A + B:")
print(C)

# Imprime las raíces cuadradas (nótese que los resultados serán flotantes)
print("\nRaíz cuadrada de los elementos de A:")
print(D)
```

***Indexación y programación orientada a objetos***

```python
import numpy as np

# Genera una "imagen" de 5x5 píxeles con valores aleatorios entre 0 y 255 (tipo uint8)
# Los parámetros son: (mínimo, máximo, tamaño_tupla)
imagen = np.random.randint(0, 256, (5, 5))

# Región de interés (ROI): Extrae una submatriz
# Usamos slicing [fila_inicio:fila_fin, col_inicio:col_fin]
# Nota: imagen[1:5, 1:5] toma desde la fila/columna 1 hasta la 4 (la 5 es exclusiva)
roi = imagen[1:5, 1:5]

# Normalización: Convierte los valores de la imagen al rango [0, 1]
# Esto es muy común en redes neuronales para que el gradiente converja más rápido
imagen_norm = imagen / 255.0

# Imprime la submatriz (4x4)
print("Región de Interés (ROI):")
print(roi)

# Imprime los valores normalizados (flotantes entre 0 y 1)
print("\nImagen Normalizada:")
print(imagen_norm)

# Imprime la matriz original completa
print("\nImagen Original:")
print(imagen)
```

***Figuras y subplots***

```python
import numpy as np
import matplotlib.pyplot as plt

# Genera 100 puntos equiespaciados entre 0 y 2π (aproximadamente 6.28)
# Esto crea el eje X para nuestras funciones
x = np.linspace(0, 2 * np.pi, 100)

# Define el tamaño de la ventana de la figura (ancho: 8 pulgadas, alto: 4 pulgadas)
plt.figure(figsize=(8, 4))

# Configura una cuadrícula de subgráficos (1 fila, 2 columnas) y selecciona el primero (índice 1)
plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x)) # Dibuja la función seno
plt.title("Función seno") # Añade un título al primer gráfico

# Selecciona el segundo subgráfico de la cuadrícula (índice 2)
plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x)) # Dibuja la función coseno
plt.title("Función coseno") # Añade un título al segundo gráfico

# Despliega la ventana con ambos gráficos en pantalla
plt.show()
```

***Gestion de gráficas***

```python
import numpy as np
import matplotlib.pyplot as plt

# Genera 100 puntos equiespaciados entre -10 y 10
# Esto define el dominio (el rango de valores en el eje X)
x = np.linspace(-10, 10, 100)

# Calcula el cuadrado de cada elemento en x (operación vectorizada)
# Esto define el rango (los valores en el eje Y)
y = x**2

# Crea una nueva ventana para la figura
plt.figure()

# Grafica x contra y. El parámetro 'label' define el texto para la leyenda.
plt.plot(x, y, label="y = x²")

# Añade un título principal en la parte superior del gráfico
plt.title("Función cuadrática")

# Etiqueta el eje horizontal (eje X)
plt.xlabel("x")

# Etiqueta el eje vertical (eje Y)
plt.ylabel("y")

# Muestra la leyenda (el recuadro que identifica la línea) basada en el 'label'
plt.legend()

# Activa la cuadrícula de fondo para facilitar la lectura de los puntos
plt.grid(True)

# Despliega la gráfica final
plt.show()
```

***Cuadrado de gráficas***

```python
import numpy as np
import matplotlib.pyplot as plt

# Genera 50 puntos entre 0 y 5
x = np.linspace(0, 5, 50)

# Calcula la función exponencial (e^x) para cada punto de x
# NumPy maneja la constante de Euler de forma optimizada
y = np.exp(x)

# Crea la ventana para la gráfica
plt.figure()

# Grafica la relación entre x y el crecimiento exponencial
plt.plot(x, y)

# Configuración de textos informativos
plt.title("Crecimiento exponencial")
plt.xlabel("x")
plt.ylabel("e^x") # Nota: e^x suele ser la etiqueta estándar para el eje Y

# Guarda la gráfica como un archivo de imagen en la carpeta actual
# dpi=300 asegura una alta resolución (calidad de impresión)
plt.savefig("Grafica_Exponencial.png", dpi=300)

# Muestra la gráfica en pantalla (importante: llamar después de savefig)
plt.show()
```

***Simula un pequeño datset de caracteristicas***

```python
import numpy as np
import pandas as pd

datos = {
    "pixel_1": [120, 130, 125],
    "pixel_2": [60, 65, 70],
    "clase": [0, 1, 2]
    }

df= pd.DataFrame(datos)
print(df)
print(df.shape)
```

***Indezado y filtrado***

***Base para separar datos de entrenamiento y prueba***

```python
#Filtrar solo la clase
df_clase_1 = df[df["clase"] == 1]

#Seleccionar columnas especificas
pixeles = df.loc[:, ["pixel_1", "pixel_2"]]

print(df_clase_1)
print(pixeles)
```

***Aritmetrica y manioulacion algebraica***

***Paso tipico previo a una red neuronal***

```python
#Normalización simple
pixeles_norm = pixeles / 255.0

#Media por columna
media = pixeles.mean()

print(pixeles_norm)
print(media)
```
