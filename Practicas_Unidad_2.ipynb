***Ejercicio 2: Capas como transformaciones.***

***Objetivo: Entender capas como funciones matematicas.***

***Nota: Cada capa aplica una transformacion lineal.***

```python
# Importamos la librería NumPy
# Se usa para trabajar con arreglos y operaciones matemáticas
import numpy as np

# Definimos el vector de entrada (features)
# Representa los datos que entran a la neurona o capa
x = np.array([1.0, 2.0])

# Definimos la matriz de pesos
# Cada fila representa los pesos de una neurona
w = np.array([[0.5, -1.0],
              [1.0,  0.5]])

# Definimos el vector de sesgo (bias)
# Se suma después de la multiplicación
b = np.array([0.1, -0.2])

# Calculamos la salida
# np.dot(w, x) realiza el producto punto (multiplicación matricial)
# Luego se suma el bias b
y = np.dot(w, x) + b

# Mostramos el resultado final
print(y)
```

***Ejercicio 3: Profundidad del modelo.***

***Objetivo: Simulare una red profunda.***

***Nota: La profundidad modifica la representacion.***

```python
import numpy as np

# Definimos la función de activación ReLU
# ReLU devuelve:
# 0 si el valor es negativo
# el mismo valor si es positivo
def relu(z):
    return np.maximum(0, z)

# Vector de entrada inicial
x = np.array([1.0, 2.0])

# Repetimos el proceso 3 veces simulando 3 capas
for i in range(3):

    # np.eye(2) crea una matriz identidad 2x2:
    # [[1, 0],
    #  [0, 1]]
    # Multiplicar por la identidad no cambia el vector

    x = relu(np.dot(np.eye(2), x))

    # Mostramos la salida de cada capa
    print(f"Capa {i+1}: {x}")
```

***Ejercicio 1: Neurona desde cero.***

***Objetivo: Implementar una neurona artificial.***

```python
import numpy as np  # Importamos NumPy para operaciones matemáticas


# Vector de entrada (3 características)
x = np.array([1.0, 2.0, -1.0])

# Vector de pesos (un peso por cada entrada)
w = np.array([0.5, -0.3, 0.8])

# Bias (sesgo) — valor escalar
b = 0.1

# Calculamos la combinación lineal:
# z = w·x + b
# np.dot realiza el producto punto entre los vectores
z = np.dot(w, x) + b

# Mostramos el resultado
print("Salida lineal:", z)
```

***Ejercicio 2: Funcion de activacion.***

***Objetivo: Aplicar activacion ReLU.***

```python
# Definimos la función de activación ReLU
# ReLU devuelve:
# - 0 si el valor es negativo
# - El mismo valor si es positivo
def relu(z):
    return np.maximum(0, z)

# Aplicamos la función ReLU a la salida lineal z
a = relu(z)

# Mostramos el resultado después de la activación
print("Salida activada:", a)
```

***Ejercicio 3: Red neuronal con libreria.***

***Objetivo: Crear una red simple con keras.***

***Abstraccion directa del modelo inalambrico.***

```python
# Importamos el modelo secuencial
# Sequential permite crear una red neuronal capa por capa en orden
from tensorflow.keras.models import Sequential

# Importamos la capa totalmente conectada (fully connected)
# Dense significa que todas las neuronas están conectadas con la capa anterior
from tensorflow.keras.layers import Dense

# Creamos el modelo secuencial
model = Sequential([

    # Primera capa:
    # 8 neuronas
    # Función de activación ReLU
    # input_shape=(3,) significa que recibe 3 valores de entrada
    Dense(8, activation="relu", input_shape=(3,)),

    # Capa de salida:
    # 1 neurona
    # Activación Sigmoid (ideal para clasificación binaria)
    Dense(1, activation="sigmoid")
])

# Muestra el resumen del modelo:
# - Número de capas
# - Forma de salida
# - Número de parámetros entrenables
model.summary()
```

***Ejercicio: MLP en Python (Keras).***

***Ejemplo tipico para la clasificacion binaria.***

```python
# Importamos el modelo secuencial
from tensorflow.keras.models import Sequential

# Importamos la capa Dense (capa totalmente conectada)
from tensorflow.keras.layers import Dense


# Creamos la red neuronal usando modelo secuencial
model = Sequential([

    # Primera capa oculta
    # 16 neuronas
    # Activación ReLU
    # input_shape=(10,) significa que el modelo recibe 10 características de entrada
    Dense(16, activation="relu", input_shape=(10,)),

    # Segunda capa oculta
    # 8 neuronas
    # También usa ReLU para aprender patrones no lineales
    Dense(8, activation="relu"),

    # Capa de salida
    # 1 neurona
    # Activación Sigmoid para clasificación binaria
    Dense(1, activation="sigmoid")
])

# Muestra el resumen del modelo
model.summary()
```

***Ejercicio: Clasificacion supervisada.***

```python
# Importamos una función para generar datos de clasificación artificiales
from sklearn.datasets import make_classification

# Importamos el modelo de Red Neuronal (Perceptrón Multicapa)
from sklearn.neural_network import MLPClassifier

# Generamos un conjunto de datos sintético
# n_samples=200 → 200 muestras (filas)
# n_features=4 → 4 características por muestra
X, y = make_classification(n_samples=200, n_features=4)

# Creamos el clasificador MLP
# hidden_layer_sizes=(10, 10) → 2 capas ocultas con 10 neuronas cada una
# max_iter=500 → máximo 500 iteraciones para entrenar
clf = MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=500)

# Entrenamos el modelo con los datos generados
clf.fit(X, y)

# Evaluamos el modelo usando los mismos datos
# score() devuelve la precisión (accuracy)
print("Precisión:", clf.score(X, y))
```

***Desenso por gradiente simple. Minimizar una funcion cuadratica.***

```python
import numpy as np  # (No es obligatorio aquí, pero puede usarse en cálculos más complejos)

# Definimos la función gradiente
# grad(x) = 2x corresponde al gradiente de la función f(x) = x²
def grad(x):
  return 2 * x

# Valor inicial
x = 10.0

# Learning rate (tasa de aprendizaje)
# Controla qué tan grande es el paso en cada iteración
lr = 0.1

# Aplicamos descenso por gradiente durante 10 iteraciones
for i in range(10):

  # Fórmula del Gradient Descent:
  # x_nuevo = x_actual - lr * gradiente
  x = x - lr * grad(x)

  # Mostramos el valor actualizado
  print(f"Iteración {i+1}: x = {x}")
```

***Ejercicio 2: Backpropagation automático***

```python
# Importamos TensorFlow
import tensorflow as tf

# Definimos una variable de TensorFlow
# tf.Variable permite que TensorFlow calcule gradientes sobre ella
x = tf.Variable(3.0)

# GradientTape registra las operaciones realizadas
# para poder calcular derivadas automáticamente
with tf.GradientTape() as tape:

    # Definimos la función
    # y = x²
    y = x**2

# Calculamos la derivada de y respecto a x
# Esto es equivalente a dy/dx
dy_dx = tape.gradient(y, x)

# Mostramos el resultado como número normal (no tensor)
print(dy_dx.numpy())
```

***Ejercicio: Creación de un modelo MLP***

```python
# Importamos el modelo secuencial
from tensorflow.keras.models import Sequential

# Importamos la capa Dense
from tensorflow.keras.layers import Dense

# Creamos el modelo correctamente usando una LISTA []
model = Sequential([

    # Primera capa oculta
    # 32 neuronas
    # ReLU como activación
    # input_shape=(10,) → recibe 10 características
    Dense(32, activation="relu", input_shape=(10,)),

    # Segunda capa oculta
    # 16 neuronas
    Dense(16, activation="relu"),

    # Capa de salida
    # 1 neurona
    # Sigmoid para clasificación binaria
    Dense(1, activation="sigmoid")
])

# Compilamos el modelo
model.compile(

    # Adam → optimizador basado en descenso por gradiente
    optimizer="adam",

    # Binary Crossentropy → función de pérdida para clasificación binaria
    loss="binary_crossentropy",

    # Métrica de evaluación
    metrics=["accuracy"]
)

# Mostramos el resumen del modelo
model.summary()
```

***Ejercicio: Comparacion de activaciones***

```python
# Importamos NumPy para cálculos matemáticos
import numpy as np

# Importamos Matplotlib para graficar
import matplotlib.pyplot as plt


# Creamos 100 valores entre -5 y 5
# Estos valores serán la entrada para las funciones
x = np.linspace(-5, 5, 100)


# Definimos las funciones de activación

# ReLU → devuelve 0 si x < 0, y x si x ≥ 0
relu = np.maximum(0, x)

# Sigmoid → transforma valores a rango (0,1)
sigmoid = 1 / (1 + np.exp(-x))

# Tanh → transforma valores a rango (-1,1)
tanh = np.tanh(x)


# Graficamos cada función
plt.plot(x, relu, label="ReLU")
plt.plot(x, sigmoid, label="Sigmoid")
plt.plot(x, tanh, label="Tanh")

# Mostramos leyenda
plt.legend()

# Mostramos cuadrícula
plt.grid(True)

# Mostramos la gráfica
plt.show()
```

***Ejercicio: Guardado y carga de un modelo.***

***Objetivo: Preparar un modelo para despliegue.***

```python
# Importamos el modelo secuencial
from tensorflow.keras.models import Sequential

# Importamos la capa Dense
from tensorflow.keras.layers import Dense

# Creamos el modelo MLP
model = Sequential([

    # Primera capa oculta
    # 32 neuronas, activación ReLU
    # Recibe 10 características de entrada
    Dense(32, activation="relu", input_shape=(10,)),

    # Segunda capa oculta
    # 16 neuronas, activación ReLU
    Dense(16, activation="relu"),

    # Capa de salida
    # 1 neurona, activación Sigmoid (clasificación binaria)
    Dense(1, activation="sigmoid")
])

# Compilamos el modelo
model.compile(

    # Adam: optimizador basado en descenso por gradiente adaptativo
    optimizer="adam",

    # Binary crossentropy: función de pérdida para clasificación binaria
    loss="binary_crossentropy",

    # Métrica de evaluación
    metrics=["accuracy"]
)

# Guardamos el modelo completo (arquitectura + pesos + configuración)
# Se guarda en formato moderno de Keras (.keras)
model.save("modelo_mlp.keras")

# Importamos la función para cargar modelos
from tensorflow.keras.models import load_model

# Cargamos el modelo guardado
modelo_cargado = load_model("modelo_mlp.keras")

# Confirmación
print("Modelo cargado correctamente")
```
